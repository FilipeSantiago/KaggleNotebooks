```python
import pandas as pd
import numpy as np
import keras

import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy

from sklearn.preprocessing import OneHotEncoder
```

## Aqui o objetivo é apenas um pequeno exemplo de como usar o Keras para uma modelagem simples

Esse é um dataset famoso, e várias coisas poderiam ser analisadas, no preprocessamento, como ver quantas pessoas com o mesmo sobrenome existem, entre outras melhorias, mas como dito no primeiro artigo, esse não é nosso objetivo aqui, e portanto essa etapa não será realizada.

```python
df = pd.read_csv("train.csv")
```

Podemos observar que os dados exibidos aqui não estão preparados pra uma rede neural, temos dados faltantes e dados alfanuméricos, aqui iremos utilizar os métodos mais simples possíveis para solucionar esses problemas, mas vale lembrar que o pré-processamento dos dados é 90% do trabalho.

```python
df
```

```python
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Survived']].ffill()

X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].copy()

y = OneHotEncoder().fit_transform(df[['Survived']].copy())


X['Sex'] = X['Sex'].astype('category')
X['Embarked'] = X['Embarked'].astype('category')

X['Sex'] = X['Sex'].cat.codes
X['Embarked'] = X['Embarked'].cat.codes

X
```

Nesse ponto o dataset já pode ser passado para uma rede neural, vamos agora passar pelo passo a passo de como o fazer utilizando uma rede neural pronta, antes de começarmos nossa jornada de constuir essa rede do zero.

```python
model = Sequential()

COLUMNS_USED = 7

model.add(Dense(100, input_shape=(7,), activation='relu'))
model.add(Dense(100, activation='relu'))

model.add(Dense(2, activation='softmax'))

model.compile(optimizer=Adam(), loss=BinaryCrossentropy())
```

```python
model.fit(X.values, y.A, epochs=100, batch_size=16)
```

```python
from sklearn.metrics import accuracy_score

preds = model.predict(X)

y_pred = np.argmax(preds, axis=1)
accuracy_score(y_pred, df['Survived'].values)
```
